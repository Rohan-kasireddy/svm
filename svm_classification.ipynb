```python
#importing libraies
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python
#dividing data into x and y
x=df.drop("happy",axis=1)
y=df["happy"]
```

```python
#spliting data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#create svm model
svm_classifier=SVC(kernel="rbf",#type of boundary
C=0.1,#regularization parameter
gamma='scale',#controls the influence on data points
verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)
```

```python
#training the model
svm_classifier.fit(x_train,y_train)
```

```python
y_pred=svm_classifier.predict(x_test)
```

```python
#evaluate the model
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

linear=linear data

poly=polynomial relationship

rbf=non linear complex data

sigmoid= Neural network like behaviour

C=1

controls overfitting vs underfitting

small c-->wider margin-->underfit

large c-->tries to classify all points--->may overfit

gamma = scale

controls how far the influence of one data point reaches

High gamma-->model memorizes
low gamma -->model generalizes
